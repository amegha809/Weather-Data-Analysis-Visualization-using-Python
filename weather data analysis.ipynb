```python
#Load the weather dataset using the Pandas library and display the first few records.
import pandas as pd 
import matplotlib.pyplot as plt
data=pd.read_csv("weather data.csv")

data.head()
```

```python
#Find the average temperature of the week using Pandas.
data['Temperature'].mean()
```

```python
#Plot a simple line graph to show temperature variation over the days of the week.
plt.figure(figsize=(6,4))
plt.grid()
plt.plot(data['Day'],data['Temperature'],marker="o",linestyle="-")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```

```python
#Create a bar chart to represent daily rainfall.
plt.bar(data['Day'],data['Rainfall'])
plt.xlabel("Days")
plt.ylabel("Rainfall")
plt.show()
#FRIDAY got the highest rainfall
```

```python
#Plot a histogram to show the distribution of temperature values.
plt.hist(data['Temperature'],bins=4)
plt.show()
```

```python
#Create a scatter plot to analyze the relationship between humidity and temperature.
#From the scatter plot, determine whether humidity and temperature are correlated.
plt.scatter(data['Temperature'],data['Humidity'])
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.show()

# Scatter plot shows a negative correlation between temperature and humidity as when temperature increases humidity decreases 
```

```python
#Plot multiple lines on the same graph to compare temperature and humidity across days.
#Add a legend to distinguish between different weather parameters.
plt.plot(data['Day'],data['Humidity'],label="Humidity",color="green",marker="o")
plt.plot(data['Day'],data['Temperature'],label="Temperature",color="blue",marker="v")
plt.legend()
plt.show()
```

```python
#Categorize days into rainy and non-rainy days and represent them using a pie chart.
#Display percentage values on the pie chart.
rainy=len(data[data['Rainfall']>0])
non_rainy=len(data[data['Rainfall']==0])
plt.pie([rainy,non_rainy],labels=['rainy','non_rainy'],autopct='%1.1f%%')
plt.show()
```

```python
#Create a single figure containing multiple subplots to visualize:
#Temperature trend
#Rainfall bar chart
#Humidity vs temperature scatter plot
#Temperature histogram
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
plt.plot(data['Day'],data['Temperature'])
plt.title("temperature trend")
plt.subplot(2,2,2)
plt.bar(data['Day'],data['Rainfall'])
plt.title("rainfall bar chart")
plt.subplot(2,2,3)
plt.scatter(data['Temperature'],data['Humidity'])
plt.title("humidity vs temperature")
plt.subplot(2,2,4)
plt.hist(data['Temperature'],bins=3)
plt.show()

# Save the combined subplot dashboard as a single image.
plt.savefig("Subplot.png")
print("image saved")
```

```python
# Use Pandas directly to generate a line plot of temperature data.
data['Temperature'].plot()
plt.show()
# Save the temperature trend plot as an image file.
plt.savefig("temp_analysis.png")
print("saved")
```
